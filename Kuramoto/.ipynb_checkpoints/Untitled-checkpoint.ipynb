```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# Single harmonic oscillator

```python
freq = 2.  # Hz
omega = 2. * np.pi * freq

def f(X, t):
    return np.array([X[1], -omega**2 * X[0]])
```

```python
%%time

tmax = 1
Nt = 1000

tv = np.linspace(0., tmax, Nt)
X = odeint(f, [1., 0.], tv)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(tv, X[:, 0], "-", label = "ODEint")
plt.xlabel("Time, $t$ [s]")
plt.ylabel("Amplitude, $x$ [m]")
plt.show()
```

# Phase model oscillator 

```python
fs = tmax / Nt

def fp(phi, t):
    return np.array([omega/fs, phi[1]])
```

```python
%%time

Phi = odeint(fp, [np.pi/2, 0.], tv)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(tv, np.sin(Phi[:, 0]), "-", label = "ODEint")
plt.xlabel("Time, $t$ [s]")
plt.ylabel("Amplitude, $x$ [m]")
plt.show()
```
